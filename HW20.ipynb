```python
class node():
    def __init__(self, val):
        self.val = val
        
    def __str__(self):
        return str(self.val)
    
class edge():
    def __init__(self, src, des, dis):
        self.src = src
        self.des = des
        self.dis = dis

class Digraph():
    def __init__(self):
        self.nodes = []
        self.edges = {}
        
    def add_node(self, n):
        if n not in self.nodes:
            self.nodes.append(n)
            
    def add_edge(self, e):
        if e.src in self.nodes or e.des in self.nodes:
            if e.src in self.edges:
                self.edges[e.src].append((e.des, e.dis))
            else:
                self.edges[e.src] = [(e.des, e.dis)]
        else:
            print('ERROR: node does not exist!')
G = Digraph()

# YOUR CODE HERE
A=node('A')
B=node('B')
C=node('C')
D=node('D')
E=node('E')
F=node('F')
Q=[A,B,C,D,E,F]
e1=edge(A,B,7)
e2=edge(A,C,9)
e3=edge(A,F,14)
e4=edge(B,C,10)
e5=edge(B,D,15)
e6=edge(C,D,11)
e7=edge(C,F,2)
e8=edge(D,E,6)
e9=edge(F,E,9)
W=[e1,e2,e3,e4,e5,e6,e7,e8,e9]
for i in Q:
    G.add_node(i)
for i in W:
    G.add_edge(i)
```

```python
import math

def SP(G, s, e):
    # YOUR CODE HERE
    allpath={}
    initPath = [s]
    pathQueue = [initPath]
    while len(pathQueue) != 0:
        #Get and remove oldest element in pathQueue
        tmpPath = pathQueue.pop(0)
        lastNode = tmpPath[-1]
        
            answer=''
            distance=0
            for i in tmpPath:
                answer+=i.val+'->'
            for i in range(len(tmpPath)-1):
                a=G.edges[tmpPath[i]]
                for j in a:
                    if j[0]==tmpPath[i+1]:
                        distance+=j[1]
            
            allpath.[(s,lastNode)]=(answer[:-2],distance)
        for i in G.edges[lastNode]:
            nextNode=i[0]
            newPath = tmpPath + [nextNode]
            pathQueue.append(newPath)
    if (s,e)in allpath.key():
        return 
```
